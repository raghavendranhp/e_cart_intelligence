# Data Exploration, Cleaning, and Feature Engineering
This notebook covers the initial loading, inspection, and transformation of the raw E-Cart datasets. The processed outputs will be used for modeling.

```python
import pandas as pd
import numpy as np
```

```python
#load datasets from raw folder
df_products = pd.read_csv('../data/raw/products.csv')
df_users = pd.read_csv('../data/raw/users.csv')
df_orders = pd.read_csv('../data/raw/orders.csv')
df_views = pd.read_csv('../data/raw/product_views.csv')
```

## 1. Data Inspection
Reviewing the shape, schema, and sample records of each dataset to understand the underlying structure.

```python
#inspect products dataset
print("products dataset info:")
df_products.info()
display(df_products.head())
display(df_products.describe(include='all'))
```

```python
#inspect users dataset
print("\nusers dataset info:")
df_users.info()
display(df_users.head())
display(df_users.describe(include='all'))
```

```python
#inspect orders dataset
print("\norders dataset info:")
df_orders.info()
display(df_orders.head())
display(df_orders.describe(include='all'))
```

```python
#inspect product views dataset
print("\nproduct views dataset info:")
df_views.info()
display(df_views.head())
display(df_views.describe(include='all'))
```

## 2. Data Cleaning & Validation
Checking for missing values and duplicated records across all datasets to ensure data integrity.

```python
#check for missing values
print("missing values check:")
print("products:\n", df_products.isnull().sum())
print("users:\n", df_users.isnull().sum())
print("orders:\n", df_orders.isnull().sum())
print("views:\n", df_views.isnull().sum())
```

```python
#check for duplicate rows
print("\nduplicate rows check:")
print("products duplicates:", df_products.duplicated().sum())
print("users duplicates:", df_users.duplicated().sum())
print("orders duplicates:", df_orders.duplicated().sum())
print("views duplicates:", df_views.duplicated().sum())
```

```python
#handle datetime conversions immediately
df_orders['order_date'] = pd.to_datetime(df_orders['order_date'])
df_views['view_time'] = pd.to_datetime(df_views['view_time'])
```

## 3. Merging Data for Comprehensive Analysis
Joining tables to calculate total revenue, category distributions, and regional insights.

```python
#merge orders with products to get price and category details
df_order_details = pd.merge(df_orders, df_products, on='product_id', how='left')

#merge with users to get demographics
df_full_data = pd.merge(df_order_details, df_users, on='user_id', how='left')

#calculate total price per order line
df_full_data['total_price'] = df_full_data['quantity'] * df_full_data['price']

display(df_full_data.head())
```

## 4. Feature Engineering
Creating specific datasets required for the Recommendation System, Demand Prediction Model, and Customer Segmentation.

```python
#feature engineering for recommendation system (implicit feedback)
#group views by user and product
view_counts = df_views.groupby(['user_id', 'product_id']).size().reset_index(name='view_count')

#group orders by user and product
order_counts = df_orders.groupby(['user_id', 'product_id'])['quantity'].sum().reset_index(name='purchase_count')
```

```python
#merge interactions
interactions = pd.merge(view_counts, order_counts, on=['user_id', 'product_id'], how='outer').fillna(0)

#calculate an implicit rating (e.g., 1 point per view, 5 points per purchase)
interactions['interaction_score'] = (interactions['view_count'] * 1) + (interactions['purchase_count'] * 5)
```

```python
#feature engineering for demand prediction (time series features)
#extract date components
df_orders['day_of_week'] = df_orders['order_date'].dt.dayofweek
df_orders['week_of_year'] = df_orders['order_date'].dt.isocalendar().week
df_orders['month'] = df_orders['order_date'].dt.month
df_orders['year'] = df_orders['order_date'].dt.year

#aggregate daily demand per product
daily_demand = df_orders.groupby(['product_id', 'order_date', 'day_of_week', 'month']).agg({'quantity': 'sum'}).reset_index()
daily_demand.rename(columns={'quantity': 'total_daily_quantity'}, inplace=True)
```

```python
#feature engineering for customer segmentation (rfm analysis)
#calculate recency, frequency, and monetary metrics per user
current_date = df_full_data['order_date'].max() + pd.Timedelta(days=1)

customer_segments = df_full_data.groupby('user_id').agg(
    recency=('order_date', lambda x: (current_date - x.max()).days),
    frequency=('order_id', 'count'),
    monetary=('total_price', 'sum')
).reset_index()
```

## 5. Exporting Processed Data
Saving the cleaned and engineered datasets to the processed directory for use in downstream notebooks.

```python
#save all processed and engineered datasets
df_products.to_csv('../data/processed/cleaned_products.csv', index=False)
df_users.to_csv('../data/processed/cleaned_users.csv', index=False)
df_orders.to_csv('../data/processed/cleaned_orders.csv', index=False)
interactions.to_csv('../data/processed/user_item_interactions.csv', index=False)
daily_demand.to_csv('../data/processed/daily_product_demand.csv', index=False)
customer_segments.to_csv('../data/processed/customer_rfm_segments.csv', index=False)

print("data cleaning and feature engineering complete. all files saved to processed directory.")
```
